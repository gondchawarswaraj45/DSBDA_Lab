```python
z_scores = stats.zscore(df['StudentID'])
```

```python
print(z_scores)
```

```python
abs_z_scores = npy.abs(z_scores)
```

```python
abs_z_scores
```

```python
outliers = (z_scores > 1.73)
```

```python
print(outliers)
```

```python
outliers = outliers[:len(df)]
df = df[~outliers]
```
