```python
import pandas as pd

data = {
    "YearsExperience": [1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.5, 3.7],
    "Salary": [39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189]
}

df = pd.DataFrame(data)
df.to_csv("Salary_Data.csv", index=False)

print("Salary_Data.csv created!")
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
grouped_stats = df.groupby('YearsExperience')['Salary'].mean()
print(grouped_stats)
```

```python
print(float(grouped_stats.median()))
```

```python
print(float(grouped_stats.min()))
```

```python
grouped_stats.std()
```

```python
grouped_stats.agg(['mean'])
```

```python
grouped_stats.agg(['mean','median','min','max','std'])
```
